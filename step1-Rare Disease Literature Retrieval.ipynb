```python
def cal_time(time):
    time = time.split('/')
    year = int(time[0])
    month = int(time[1]) - 12
    day = int(time[2])
    if month < 1:
        month += 12
        year -= 1
    time = datetime.datetime(year=year, month=month, day=1).strftime('%Y/%m/%d')
    return time
```
