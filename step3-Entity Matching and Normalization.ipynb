```python
# Open the file and read all lines
def eram_disease(file_name,disease_names_eram):
    with open(file_name, 'r') as file:
        lines = file.readlines()
    # Extract disease names (first column) from each line, skipping the header
    for line in lines[1:]:  # Skip the header line
        match = re.search(r'\t|[A-Z]', line)
        if match:
            index = match.start()
            parts = line[:index]
        if parts:  # Ensure the line is not empty
            disease_name = parts
            disease_names_eram.add(re.sub(r'[^\w\s]', '', disease_name).lower())
    return disease_names_eram
```

```python
import xml.etree.ElementTree as ET

# Parse the XML file
tree = ET.parse('.\\Validation.\\ORPHNET.\\en_product6.xml')
root = tree.getroot()

# Create a set to store unique gene names
gene_set = set()

# Iterate through all Disorder elements
for disorder in root.findall('.//Disorder'):
    # Iterate through all DisorderGeneAssociation elements
    for dga in disorder.findall('.//DisorderGeneAssociation'):
        # Find the Gene element
        gene = dga.find('Gene')
        if gene is not None:
            # Get the Symbol (primary gene name)
            symbol = gene.find('Symbol')
            if symbol is not None and symbol.text:
                gene_set.add(symbol.text.lower())
            
            # Also get all synonyms
            synonym_list = gene.find('SynonymList')
            if synonym_list is not None:
                for synonym in synonym_list.findall('Synonym'):
                    if synonym.text:
                        gene_set.add(synonym.text.lower())

# Print the set of genes
print(f"\nTotal unique gene names: {len(gene_set)}")
```

```python
# Read the file content
def get_eram_genes(file_name,genes):
    with open(file_name) as file:
        content = file.read()
    
    # Split the content into lines
    lines = content.split('\n')
    # Initialize a set to store unique genes (using a set to avoid duplicates)
    parts=[]
    entries=[]
    # Skip the header line and process each line
    for line in lines[1:]:
        if not line:
            continue  # Skip empty lines
        match = re.search(r'\t|[A-Z]', line)
        if match:
            index = match.start()
            parts = line[index:]
            # Split gene entries by '#' to handle multiple entries per line
            entries = parts.split('#')
        for entry in entries:
            # Split each entry by '|' to get the gene name
            gene = entry.split('|')[0]
            genes.add(re.sub('\t', '', gene).lower())
    return genes
```
