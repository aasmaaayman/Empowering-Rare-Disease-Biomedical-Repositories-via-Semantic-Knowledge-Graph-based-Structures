```python
from pathlib import Path
from owlready2 import World

def load_ontology_local(path):
    """
    Safely load a local OWL ontology file (Windows-friendly, duplicate-safe).
    Uses an isolated Owlready2 World to prevent UNIQUE constraint errors.
    """
    path = Path(path).resolve()
    if not path.exists():
        raise FileNotFoundError(f" Ontology file not found: {path}")
    
    print(f" Loading ontology safely: {path}")

    # Create a separate Owlready2 "world" (temporary SQLite DB)
    world = World()
    onto = world.get_ontology(str(path))
    try:
        onto.load()
        print(f"   Classes loaded: {len(list(onto.classes()))}")
    except Exception as e:
        print(f" Warning while loading {path.name}: {e}")
    
    return onto
```

```python
# ==============================================================
# 3. LABEL INDEX BUILDING
# ==============================================================

def build_label_index(onto):
    label_map = {}
    for c in onto.classes():
        labels = set()
        if hasattr(c, "label"):
            for lab in c.label:
                labels.add(str(lab).lower())
        for prop in ("hasExactSynonym", "hasRelatedSynonym", "altLabel", "synonym"):
            if hasattr(c, prop):
                for s in getattr(c, prop):
                    labels.add(str(s).lower())
        for lab in labels:
            label_map[lab] = c.iri
    return label_map
```

```python
# ==============================================================
# 4. PUBTATOR PARSING
# ==============================================================
def fetch_pubtator_json_for_pmid(pmid):
    """
    Fetch PubTator3 BioC JSON annotations for a given PMID.
    """
    url = f"https://www.ncbi.nlm.nih.gov/research/pubtator3-api/publications/export/biocjson?pmids={pmid}"
    try:
        r = requests.get(url, timeout=30)
        r.raise_for_status()
        data = r.json()
        if isinstance(data, dict) and "documents" in data:
            return data["documents"][0]
        else:
            return data
    except Exception as e:
        print(f" fetch error {pmid}: {e}")
        return None
```

```python
# ==============================================================
# 5. PUBTATOR PARSER
# ==============================================================
def parse_pubtator_doc(doc):
    """
    Parse a PubTator3 document in the format:
    {'PubTator3': [ { 'id': '...', 'passages': [...], 'annotations': [...], ... } ]}
    """
    # --- handle wrapper ---
    if not doc:
        return {"pmid": None, "text": "", "sentences": [], "annotations": []}

    # Extract the inner document object
    if "PubTator3" in doc and isinstance(doc["PubTator3"], list) and len(doc["PubTator3"]) > 0:
        doc = doc["PubTator3"][0]
    else:
        return {"pmid": None, "text": "", "sentences": [], "annotations": []}

    # --- extract content ---
    pmid = str(doc.get("pmid") or doc.get("id"))
    text = ""
    sentences = []
    anns = []

    for p in doc.get("passages", []):
        passage_text = p.get("text", "")
        start = p.get("offset", 0)
        if passage_text:
            sentences.append({
                "start": start,
                "end": start + len(passage_text),
                "text": passage_text
            })
            text += passage_text + " "

        # Parse each annotation in the passage
        for ann in p.get("annotations", []):
            inf = ann.get("infons", {})
            loc = ann.get("locations", [{}])[0]
            anns.append({
                "type": inf.get("type") or inf.get("biotype"),
                "text": ann.get("text"),
                "start": loc.get("offset"),
                "end": loc.get("offset", 0) + loc.get("length", 0),
                "id": inf.get("identifier") or inf.get("normalized_id") or ann.get("id"),
                "normalized": inf.get("normalized_id") or inf.get("normalized"),
                "database": inf.get("database"),
                "name": inf.get("name")
            })

    return {
        "pmid": pmid,
        "text": text.strip(),
        "sentences": sentences,
        "annotations": anns
    }
```

```python
# ==============================================================
# 6. NORMALIZATION
# ==============================================================

from Levenshtein import distance as levenshtein_distance

def normalize_text(text, label_map, max_dist=1):
    if not text:
        return None
    t = text.lower().strip()
    if t in label_map:
        return label_map[t]
    for label in label_map.keys():
        if abs(len(label) - len(t)) < 6 and levenshtein_distance(label, t) <= max_dist:
            return label_map[label]
    return None
def find_related_entities(entity_iri, ontology, keywords):
    """
    Search ontology for related classes based on property name keywords.
    """
    related = set()
    cls = ontology.search_one(iri=entity_iri)
    if not cls:
        return []
    for prop in ontology.object_properties():
        if any(k.lower() in prop.name.lower() for k in keywords):
            try:
                for val in getattr(cls, prop.name, []):
                    related.add(val)
            except Exception:
                continue
    return list(related)
```

```python
import pandas as pd
import json
from pronto import Ontology
hpo_dict = {}
for term in hpo.terms():
    if term.id.startswith("HP:"):
        label = term.name
        synonyms = [syn.description for syn in term.synonyms]
        parents = list(term.superclasses(distance=1))
        parents = [p for p in parents if p.id != term.id]
        parent_id = parents[0].id if parents else None
        parent_label = parents[0].name if parents else None
        xrefs = [str(x).split("'")[1] for x in term.xrefs] if term.xrefs else []

        hpo_dict[term.id] = {
            "HPO_Label": label,
            "Synonyms": synonyms,
            "Parent_Term": parent_id,
            "Parent_Label": parent_label,
            "Cross_References": xrefs
        }

print("HPO ontology loaded.")

# -----------------------------
# 2 Load phenotype annotations
# -----------------------------
phenotype_df = pd.read_csv(
    "phenotype.hpoa",
    sep="\t",
    comment="#",
    names=["DB_ID", "Disease_Name", "Qualifier", "HPO_ID", "Reference",
           "Evidence", "Onset", "Frequency", "Sex", "Modifier", "Aspect",
           "Biocuration"],
    dtype=str,
    low_memory=False
)
phenotype_df = phenotype_df[["DB_ID", "HPO_ID", "Frequency", "Evidence",
                             "Onset", "Modifier", "Qualifier"]]
phenotype_df["Frequency"] = phenotype_df["Frequency"].astype(float, errors='ignore')

phenotype_lookup = {
    (row['DB_ID'], row['HPO_ID']): row.to_dict()
    for _, row in phenotype_df.iterrows()
}

print("Phenotype annotations loaded.")

# -----------------------------
# 3 Load MAXO annotations
# -----------------------------
maxo_df = pd.read_csv("maxo-annotations.tsv", sep="\t", dtype=str, low_memory=False)
maxo_df = maxo_df[["disease_id", "maxo_id", "maxo_name", "hpo_id",
                   "relation", "evidence", "extension_id", "extension_name"]]
maxo_lookup = {
    (row['disease_id'], row['hpo_id']): row.to_dict()
    for _, row in maxo_df.iterrows()
}

print("MAXO annotations loaded.")

# -----------------------------
# 4 Load gene-phenotype associations (updated columns)
# ncbi_gene_id | gene_symbol | hpo_id | hpo_name | frequency | disease_id
# -----------------------------
gene_df = pd.read_csv("genes_to_phenotype.txt", sep="\t", dtype=str)

# Build HPO_ID  list of genes (gene symbols)
hpo_to_genes = gene_df.groupby("hpo_id")["gene_symbol"].apply(list).to_dict()
print("Gene-phenotype associations loaded.")

# -----------------------------
# 5 Load disease CSV
# -----------------------------
df = pd.read_csv("disease_phenotypes_fast.csv")  # Columns: MeSH_ID, MONDO_ID, Phenotype_ID

# -----------------------------
# 6 Enrich each disease-phenotype row
# -----------------------------
enriched_records = []

for _, row in df.iterrows():
    hpo_id = row['Phenotype_ID']
    disease_id = row['MONDO_ID']

    record = {
        "MeSH_ID": row['MeSH_ID'],
        "MONDO_ID": disease_id,
        "HPO_ID": hpo_id
    }

    # HPO ontology info
    if hpo_id in hpo_dict:
        record.update(hpo_dict[hpo_id])

    # Phenotype annotations
    pheno_ann = phenotype_lookup.get((disease_id, hpo_id))
    if pheno_ann:
        record.update({
            "Frequency": pheno_ann["Frequency"],
            "Evidence": pheno_ann["Evidence"],
            "Onset": pheno_ann["Onset"],
            "Modifier": pheno_ann["Modifier"],
            "Abnormality_Type": "Present",
            "Negated": False if pheno_ann["Qualifier"] in ["", "Present"] else True
        })
    else:
        record.update({
            "Frequency": None,
            "Evidence": None,
            "Onset": None,
            "Modifier": None,
            "Abnormality_Type": "Present",
            "Negated": False
        })

    # MAXO annotation info
    maxo_ann = maxo_lookup.get((disease_id, hpo_id))
    if maxo_ann:
        record.update({
            "MAXO_ID": maxo_ann["maxo_id"],
            "MAXO_Name": maxo_ann["maxo_name"],
            "Relation": maxo_ann["relation"],
            "MAXO_Evidence": maxo_ann["evidence"],
            "Extension_ID": maxo_ann["extension_id"],
            "Extension_Name": maxo_ann["extension_name"]
        })

    # Related genes from updated HPO-to-gene mapping
    record["Related_Genes"] = hpo_to_genes.get(hpo_id, [])

    # Confidence
    record["Confidence"] = "High" if record["Evidence"] or record.get("MAXO_Evidence") else "Medium"

    enriched_records.append(record)

# -----------------------------
# 7 Save fully enriched JSON
# -----------------------------
with open("fully_enriched_diseases.json", "w", encoding="utf-8") as f:
    json.dump(enriched_records, f, indent=2, ensure_ascii=False)

print("Saved fully enriched data to fully_enriched_diseases.json")
```

```python
from rdflib import Graph, RDFS, Namespace
import re
from collections import defaultdict

g = Graph()
g.parse("mondo-rare.owl")

OBO = Namespace("http://www.geneontology.org/formats/oboInOwl#")
IAO = Namespace("http://purl.obolibrary.org/obo/IAO_")

mondo = defaultdict(lambda: {"label": None, "synonyms": set(), "xrefs": set()})

def is_mondo(uri):
    return "MONDO_" in str(uri)

def get_id(uri):
    return re.sub(r".*MONDO_", "MONDO_", str(uri))

# Extract labels
for s, p, o in g.triples((None, RDFS.label, None)):
    if is_mondo(s):
        mondo_id = get_id(s)
        mondo[mondo_id]["label"] = str(o).lower()

# Extract synonyms
for s, p, o in g.triples((None, OBO.hasExactSynonym, None)):
    if is_mondo(s):
        mondo_id = get_id(s)
        mondo[mondo_id]["synonyms"].add(str(o).lower())

# Extract xrefs
for s, p, o in g.triples((None, OBO.hasDbXref, None)):
    if is_mondo(s):
        mondo_id = get_id(s)
        mondo[mondo_id]["xrefs"].add(str(o))
```

```python
#!/usr/bin/env python3
"""
Mondo Rare Disease Enricher for PubTator3 Output
Enriches diseases from PubTator3 with Mondo rare disease ontology
Output: JSON format
"""

import json
import pandas as pd
from owlready2 import get_ontology, default_world, OwlReadyOntologyParsingError
import re
from datetime import datetime
from typing import Dict, List, Optional, Any
import os

class MondoRareEnricher:
    def __init__(self, mondo_owl_path: str = "mondo-rare.owl"):
        """
        Initialize the Mondo enricher with mondo-rare.owl
        
        Args:
            mondo_owl_path: Path to mondo-rare.owl file
        """
        print(f" Loading Mondo rare disease ontology from {mondo_owl_path}...")
        
        # Check if file exists
        if not os.path.exists(mondo_owl_path):
            print(f" Error: File {mondo_owl_path} not found!")
            print(f"   Current directory: {os.getcwd()}")
            print(f"   Files in directory: {os.listdir('.')}")
            raise FileNotFoundError(f"Mondo OWL file not found: {mondo_owl_path}")
        
        # Check file size
        file_size = os.path.getsize(mondo_owl_path)
        print(f"   File size: {file_size / (1024*1024):.2f} MB")
        
        if file_size == 0:
            print(" Error: File is empty!")
            raise ValueError("Mondo OWL file is empty")
        
        try:
            # Try with different parsing options if needed
            default_world.set_backend(filename="pydeds")  # Use pydeds backend for better RDF/XML parsing
            
            # Load the ontology with verbose output
            print("   Parsing ontology (this may take a moment)...")
            self.onto = get_ontology(mondo_owl_path).load()
            
            print(f" Successfully loaded Mondo ontology")
            print(f"   Number of classes: {len(list(self.onto.classes()))}")
            print(f"   Number of properties: {len(list(self.onto.properties()))}")
            
            # Build mappings
            print("   Building mappings...")
            self.xref_to_mondo = self._build_xref_mapping()
            self.mondo_cache = {}
            
            # Load mondo hierarchy for parent relationships
            self.hierarchy = self._extract_hierarchy()
            
            print(f" Built mappings for {len(self.xref_to_mondo)} external identifiers")
            
        except OwlReadyOntologyParsingError as e:
            print(f" RDF/XML parsing error: {e}")
            print("This might be due to malformed XML in the OWL file.")
            print("Suggestions:")
            print("1. Download a fresh copy of mondo-rare.owl")
            print("2. Try a smaller subset of Mondo if available")
            print("3. Check if the file is corrupted")
            raise
        except Exception as e:
            print(f" Error loading ontology: {e}")
            print(f"Error type: {type(e)}")
            import traceback
            traceback.print_exc()
            raise
    
    def _build_xref_mapping(self) -> Dict[str, str]:
        """Build comprehensive mapping from external IDs to MONDO IDs"""
        xref_map = {}
        count = 0
        
        for cls in self.onto.classes():
            # Get MONDO ID in format MONDO:XXXXXXX
            mondo_id = cls.iri.split('/')[-1].replace('_', ':')
            
            # Map all external references
            if hasattr(cls, 'hasDbXref'):
                for xref in cls.hasDbXref:
                    xref_str = str(xref)
                    # Store mapping
                    xref_map[xref_str] = mondo_id
                    
                    # Also create normalized versions
                    if xref_str.startswith('MESH:'):
                        xref_map[xref_str.replace('MESH:', 'MeSH:')] = mondo_id
                        xref_map[xref_str.replace('MESH:', 'mesh:')] = mondo_id
                    elif xref_str.startswith('MeSH:'):
                        xref_map[xref_str.replace('MeSH:', 'MESH:')] = mondo_id
                        xref_map[xref_str.replace('MeSH:', 'mesh:')] = mondo_id
                    elif xref_str.startswith('UMLS:'):
                        xref_map[xref_str.upper()] = mondo_id
                        xref_map[xref_str.lower()] = mondo_id
            
            # Map from labels and synonyms (for name matching)
            if cls.label:
                label = cls.label.first().lower()
                xref_map[f"label:{label}"] = mondo_id
            
            if hasattr(cls, 'hasExactSynonym'):
                for syn in cls.hasExactSynonym:
                    xref_map[f"synonym:{str(syn).lower()}"] = mondo_id
            
            count += 1
            if count % 1000 == 0:
                print(f"   Processed {count} classes...")
        
        return xref_map
    
    def _extract_hierarchy(self) -> Dict[str, List[str]]:
        """Extract parent-child relationships from ontology"""
        hierarchy = {}
        count = 0
        
        for cls in self.onto.classes():
            mondo_id = cls.iri.split('/')[-1].replace('_', ':')
            parents = []
            
            # Get parent classes (is_a relationships)
            for parent in cls.is_a:
                if hasattr(parent, 'iri'):
                    parent_id = parent.iri.split('/')[-1].replace('_', ':')
                    parents.append(parent_id)
            
            hierarchy[mondo_id] = parents
            
            count += 1
            if count % 1000 == 0:
                print(f"   Processed hierarchy for {count} classes...")
        
        return hierarchy
    
    def _get_mesh_id(self, identifier: str) -> str:
        """Extract and normalize MeSH ID from identifier"""
        # Handle different formats
        if identifier.startswith('mesh:'):
            return identifier.replace('mesh:', '').upper()
        elif identifier.startswith('MeSH:'):
            return identifier.replace('MeSH:', '').upper()
        elif identifier.startswith('MESH:'):
            return identifier.replace('MESH:', '').upper()
        else:
            # Assume it's already a MeSH ID
            return identifier.upper()
    
    def _find_mondo_by_mesh(self, mesh_id: str) -> Optional[str]:
        """Find MONDO ID by MeSH ID with multiple lookup strategies"""
        # Try exact MeSH ID matches
        lookup_patterns = [
            f"MeSH:{mesh_id}",
            f"MESH:{mesh_id}",
            f"mesh:{mesh_id.lower()}",
            mesh_id,  # Sometimes xrefs are just the ID
        ]
        
        for pattern in lookup_patterns:
            if pattern in self.xref_to_mondo:
                return self.xref_to_mondo[pattern]
        
        return None
    
    def _find_mondo_by_name(self, disease_name: str) -> Optional[str]:
        """Find MONDO ID by disease name using fuzzy matching"""
        normalized_name = disease_name.lower().strip()
        
        # Try exact label match
        label_key = f"label:{normalized_name}"
        if label_key in self.xref_to_mondo:
            return self.xref_to_mondo[label_key]
        
        # Try synonym match
        synonym_key = f"synonym:{normalized_name}"
        if synonym_key in self.xref_to_mondo:
            return self.xref_to_mondo[synonym_key]
        
        # Try partial matching
        for key in self.xref_to_mondo:
            if key.startswith('label:') and normalized_name in key:
                return self.xref_to_mondo[key]
            if key.startswith('synonym:') and normalized_name in key:
                return self.xref_to_mondo[key]
        
        return None
    
    def _get_class_info(self, mondo_id: str) -> Dict[str, Any]:
        """Get detailed information for a MONDO ID"""
        if mondo_id in self.mondo_cache:
            return self.mondo_cache[mondo_id]
        
        # Find the class in ontology
        cls = self.onto.search_one(iri=f"*{mondo_id.replace(':', '_')}")
        if not cls:
            return {}
        
        # Extract information
        info = {
            'mondo_id': mondo_id,
            'label': cls.label.first() if cls.label else '',
            'definition': '',
            'synonyms': [],
            'xrefs': [],
            'parents': self.hierarchy.get(mondo_id, []),
            'all_parents': []
        }
        
        # Get definition
        if hasattr(cls, 'hasDefinition') and cls.hasDefinition:
            info['definition'] = str(cls.hasDefinition.first())
        
        # Get synonyms
        if hasattr(cls, 'hasExactSynonym'):
            info['synonyms'] = [str(syn) for syn in cls.hasExactSynonym]
        if hasattr(cls, 'hasRelatedSynonym'):
            info['synonyms'].extend([str(syn) for syn in cls.hasRelatedSynonym])
        
        # Get external references
        if hasattr(cls, 'hasDbXref'):
            info['xrefs'] = [str(xref) for xref in cls.hasDbXref]
        
        # Get all ancestors (transitive closure)
        all_parents = set()
        to_process = info['parents'].copy()
        while to_process:
            parent = to_process.pop()
            if parent not in all_parents:
                all_parents.add(parent)
                to_process.extend(self.hierarchy.get(parent, []))
        info['all_parents'] = list(all_parents)
        
        # Cache the results
        self.mondo_cache[mondo_id] = info
        return info
    
    def enrich_disease(self, identifier: str, disease_name: str) -> Dict[str, Any]:
        """
        Enrich a single disease from PubTator3
        
        Args:
            identifier: MeSH identifier from PubTator3 (e.g., 'mesh:d009456')
            disease_name: Disease name from PubTator3
            
        Returns:
            Dictionary with enriched information
        """
        # Start with base structure
        result = {
            'pubtator_data': {
                'identifier': identifier,
                'disease_name': disease_name,
                'mesh_id': self._get_mesh_id(identifier)
            },
            'mondo_enrichment': {
                'found_in_mondo': False,
                'mapping_method': 'not_found'
            },
            'enriched_data': {},
            'timestamp': datetime.now().isoformat()
        }
        
        # Try to find MONDO ID by MeSH ID
        mesh_id = result['pubtator_data']['mesh_id']
        mondo_id = self._find_mondo_by_mesh(mesh_id)
        mapping_method = 'mesh_id_direct'
        
        # If not found by MeSH, try by name
        if not mondo_id:
            mondo_id = self._find_mondo_by_name(disease_name)
            mapping_method = 'name_match' if mondo_id else 'not_found'
        
        # If found, get detailed information
        if mondo_id:
            class_info = self._get_class_info(mondo_id)
            
            result['mondo_enrichment'].update({
                'found_in_mondo': True,
                'mapping_method': mapping_method,
                'mondo_id': mondo_id
            })
            
            # Add enriched data
            result['enriched_data'] = {
                'standard_name': class_info['label'],
                'definition': class_info['definition'],
                'synonyms': class_info['synonyms'],
                'external_references': class_info['xrefs'],
                'parent_classes': class_info['parents'],
                'all_ancestor_classes': class_info['all_parents'],
                'is_rare_disease': True,
                'confidence_score': 'high' if mapping_method == 'mesh_id_direct' else 'medium'
            }
            
            # Extract key external references by type
            key_refs = {
                'omim': [],
                'orphanet': [],
                'umls': [],
                'doid': [],
                'gard': [],
                'other': []
            }
            
            for xref in class_info['xrefs']:
                if xref.startswith('OMIM:'):
                    key_refs['omim'].append(xref)
                elif xref.startswith('Orphanet:') or xref.startswith('ORPHA:'):
                    key_refs['orphanet'].append(xref)
                elif xref.startswith('UMLS:'):
                    key_refs['umls'].append(xref)
                elif xref.startswith('DOID:'):
                    key_refs['doid'].append(xref)
                elif xref.startswith('GARD:'):
                    key_refs['gard'].append(xref)
                else:
                    key_refs['other'].append(xref)
            
            result['enriched_data']['key_external_references'] = key_refs
            
            # Add hierarchy path as string for readability
            if class_info['parents']:
                parent_labels = []
                for parent_id in class_info['parents'][:3]:  # Show up to 3 parents
                    parent_info = self._get_class_info(parent_id)
                    parent_labels.append(parent_info.get('label', parent_id))
                result['enriched_data']['hierarchy_path'] = '  '.join(parent_labels)
        
        else:
            # Disease not found in MONDO rare diseases
            result['enriched_data'] = {
                'is_rare_disease': False,
                'notes': 'Disease not found in MONDO rare disease ontology. This may indicate: 1) Not a rare disease, 2) Different terminology used, 3) Not included in mondo-rare subset.'
            }
        
        return result


# Alternative minimal version for testing with a small dataset
class MondoTestEnricher:
    """Minimal enricher for testing when full Mondo is not available"""
    
    def enrich_disease(self, identifier: str, disease_name: str) -> Dict[str, Any]:
        """Simple enrichment for testing"""
        result = {
            'pubtator_data': {
                'identifier': identifier,
                'disease_name': disease_name,
                'mesh_id': identifier.upper().replace('MESH:', '').replace('MESH:', '')
            },
            'mondo_enrichment': {
                'found_in_mondo': False,
                'mapping_method': 'not_found'
            },
            'enriched_data': {},
            'timestamp': datetime.now().isoformat()
        }
        
        # Simple test matching
        mesh_id = result['pubtator_data']['mesh_id']
        if f"MeSH:{mesh_id}" in self.xref_to_mondo:
            mondo_id = self.xref_to_mondo[f"MeSH:{mesh_id}"]
            result['mondo_enrichment'] = {
                'found_in_mondo': True,
                'mapping_method': 'mesh_id_direct',
                'mondo_id': mondo_id
            }
            result['enriched_data'] = {
                'standard_name': disease_name,
                'is_rare_disease': True,
                'confidence_score': 'high'
            }
        
        return result


def main():
    """
    Main function to process PubTator3 diseases and enrich with Mondo
    """
    
    # Sample diseases from PubTator3 (you would replace this with your actual data)
    # This is just for testing - replace with your actual diseases_pubmed list
    
    print("=" * 60)
    print("MONDO RARE DISEASE ENRICHMENT FOR PUBTATOR3 OUTPUT")
    print("=" * 60)
    print(f"Processing {len(diseases_pubmed)} diseases from PubTator3...\n")
    
    # Try to initialize enricher with mondo-rare.owl
    enricher = None
    use_test_enricher = False
    
    try:
        enricher = MondoRareEnricher("mondo-rare.owl")
    except Exception as e:
        print(f"\n  Could not load mondo-rare.owl: {e}")
        print("Falling back to test enricher with limited data.")
        print("\nTo fix this issue:")
        print("1. Download mondo-rare.owl from: https://github.com/monarch-initiative/mondo/releases")
        print("2. Ensure it's in the same directory as this script")
        print("3. Check file permissions and integrity")
        print("\nFor now, using test data...")
        enricher = MondoTestEnricher()
        use_test_enricher = True
    
    # Enrich all diseases
    enriched_results = []
    
    for disease in diseases_pubmed:
        print(f" Enriching: {disease['Disease']} ({disease['Identifier']})")
        
        result = enricher.enrich_disease(
            identifier=disease['Identifier'],
            disease_name=disease['Disease']
        )
        
        # Display quick summary
        if result['mondo_enrichment']['found_in_mondo']:
            mondo_id = result['mondo_enrichment']['mondo_id']
            std_name = result['enriched_data'].get('standard_name', disease['Disease'])
            print(f"    Found in MONDO: {mondo_id} ({std_name})")
        else:
            print(f"    Not found in MONDO rare diseases")
        
        enriched_results.append(result)
    
    print("\n" + "=" * 60)
    print(" ENRICHMENT SUMMARY")
    print("=" * 60)
    
    # Calculate statistics
    found_count = sum(1 for r in enriched_results if r['mondo_enrichment']['found_in_mondo'])
    print(f"Total diseases processed: {len(enriched_results)}")
    print(f"Found in MONDO rare diseases: {found_count}")
    print(f"Not found in MONDO: {len(enriched_results) - found_count}")
    
    if use_test_enricher:
        print("\n  NOTE: Using test data. Results are for demonstration only.")
        print("   Download mondo-rare.owl for real enrichment.")
    
    # Prepare final JSON output
    output_data = {
        'metadata': {
            'source': 'PubTator3',
            'enrichment_source': 'test_data' if use_test_enricher else 'mondo-rare.owl',
            'processing_date': datetime.now().isoformat(),
            'total_diseases': len(enriched_results),
            'found_in_mondo': found_count,
            'not_found': len(enriched_results) - found_count,
            'note': 'Test data only' if use_test_enricher else 'Full Mondo enrichment'
        },
        'enriched_diseases': enriched_results,
        'statistics_by_mapping_method': {
            'mesh_id_direct': sum(1 for r in enriched_results 
                                if r['mondo_enrichment']['mapping_method'] == 'mesh_id_direct'),
            'name_match': sum(1 for r in enriched_results 
                            if r['mondo_enrichment']['mapping_method'] == 'name_match'),
            'not_found': sum(1 for r in enriched_results 
                           if r['mondo_enrichment']['mapping_method'] == 'not_found')
        }
    }
    
    # Save to JSON file
    output_filename = f"pubtator3_mondo_enriched_{datetime.now().strftime('%Y%m%d_%H%M%S')}.json"
    
    with open(output_filename, 'w', encoding='utf-8') as f:
        json.dump(output_data, f, indent=2, ensure_ascii=False)
    
    print(f"\n JSON output saved to: {output_filename}")
    
    # Also create a simplified CSV for quick viewing
    simplified_data = []
    for result in enriched_results:
        simple = {
            'pubtator_identifier': result['pubtator_data']['identifier'],
            'pubtator_disease_name': result['pubtator_data']['disease_name'],
            'found_in_mondo': result['mondo_enrichment']['found_in_mondo'],
            'mondo_id': result['mondo_enrichment'].get('mondo_id', ''),
            'standard_name': result['enriched_data'].get('standard_name', ''),
            'is_rare_disease': result['enriched_data'].get('is_rare_disease', False),
            'mapping_method': result['mondo_enrichment']['mapping_method']
        }
        simplified_data.append(simple)
    
    # Save simplified version
    df = pd.DataFrame(simplified_data)
    csv_filename = output_filename.replace('.json', '_simplified.csv')
    df.to_csv(csv_filename, index=False)
    print(f" Simplified CSV saved to: {csv_filename}")
    
    # Display sample of enriched data
    print("\n" + "=" * 60)
    print(" SAMPLE ENRICHED RESULTS")
    print("=" * 60)
    
    for i, result in enumerate(enriched_results[:3]):  # Show first 3
        print(f"\n{i+1}. {result['pubtator_data']['disease_name']}")
        if result['mondo_enrichment']['found_in_mondo']:
            print(f"   MONDO ID: {result['mondo_enrichment']['mondo_id']}")
            print(f"   Standard Name: {result['enriched_data']['standard_name']}")
            if result['enriched_data'].get('definition'):
                print(f"   Definition: {result['enriched_data']['definition'][:100]}...")
            if result['enriched_data'].get('key_external_references', {}).get('omim'):
                print(f"   OMIM Links: {', '.join(result['enriched_data']['key_external_references']['omim'][:2])}")
        else:
            print(f"    Not a rare disease in MONDO")
    
    return output_data


if __name__ == "__main__":
    # Run the enrichment
    print("\n Starting Mondo Rare Disease Enricher...")
    final_output = main()
    
    print("\n" + "=" * 60)
    print(" ENRICHMENT COMPLETE!")
    print("=" * 60)
    print("\nYour enriched data includes:")
    print("1. Original PubTator3 identifiers and names")
    print("2. MONDO IDs for rare diseases found")
    print("3. Standardized disease names")
    print("4. Definitions and synonyms")
    print("5. Links to OMIM, Orphanet, UMLS, etc.")
    print("6. Disease hierarchy and classification")
    print("7. Confidence scores for mappings")
    print("\nUse the JSON file for further analysis or integration!")
```
